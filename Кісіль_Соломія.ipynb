```python
import matplotlib.pyplot as plt
import numpy as np

# Визначаємо функцію параболи
def parabola(x):
    return x**2

# Генеруємо значення x
x = np.linspace(-10, 10, 400)

# Обчислюємо значення y
y = parabola(x)

# Створюємо графік
plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title('Графік параболи')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
